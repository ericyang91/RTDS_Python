```python
"""
1. Write a program that reads a text from a user, then counts and displays
how many words and how many alphanumeric letters are in the text.
"""

def word_char_counter():
    while True:
        try:
            text = input('Enter your text here: ').strip() # This makes sure no empty spaces are entered. Python regards empty spaces as characters.
            if len(text) == 0:
                raise ValueError
        except ValueError:
            print('Invalid input. Please enter your text.')
        else:
            break

    word_lst = text.split(' ')

    alphanumeric_count = 0
    for character in text:
        if character.isalnum() == True:
            alphanumeric_count += 1

    print(f"There are {len(word_lst)} words and {alphanumeric_count} alphanumeric characters in the text.")

word_char_counter()
```

```python
"""
2. Write a program that

a. reads a series of numbers that are separated by whitespace and
uses a new line to end the input, then converts the numbers in the
input string and puts them into a list.

b. sorts the numbers in the list in descending order, using the sort()
list object method.

c. sorts the numbers in the list in descending order, using the Python
built-in function sorted().

d. Write your own code to sort the numbers in the list in ascending order
without using the sort() method or sorted() function.
"""

def selection_sort(list_of_numbers): # Selection Sort: Finds the minimum (or maximum) element from the unsorted part and swaps it with the first unsorted element.
    for i in range(len(list_of_numbers)):
        for j in range(i+1, len(list_of_numbers)):
            if list_of_numbers[i] > list_of_numbers[j]:
                list_of_numbers[i], list_of_numbers[j] = list_of_numbers[j], list_of_numbers[i]
    return list_of_numbers

def number_sorting():
    while True:
        try:
            user_numbers = input('Enter your numbers separated only by whitespace: ').strip()
            if not user_numbers:
                raise ValueError('You have to enter at least one number. ')
            if '.' in user_numbers or ',' in user_numbers or ';' in user_numbers or '  ' in user_numbers: # if ',' or '.' in user_numbers won't work because ',' is a non-empty string and will always result in True.
                raise ValueError('You can separate numbers only by using a whitespace. ')
        except ValueError as e:
            print(e)
        else:
            user_numbers = user_numbers.split(' ')
            number_lst = [float(number) for number in user_numbers]
            break
    
    use_sorted = sorted(number_lst, reverse=True) # sorted() returns a new list
    number_lst.sort(reverse=True) # list.sort() modifies existing list
    use_selection_sort = selection_sort(number_lst)

    print(f'Using sorted(): {use_sorted}')
    print(f'Using list.sort(): {number_lst}')
    print(f'Using selection sort: {use_selection_sort}')

number_sorting()
```

```python
"""
Problem 5: Python Quiz Using Text Handlers

This script reads multiple-choice questions from a text file and allows the user to take a quiz by answering
a user-specified number of random questions. The responses are recorded, and the result is returned back to the user. 
It is also written to an output file.

Input:
    - Prompts the user to input the number of questions they would like to tackle (between 1 and 10).
    - Prompts the user for answers.

Output:
    - Returns whether the user's answers are correct or incorrect.
    - Prints the user's score and details of incorrect answers.
    - Writes the result to an output file.

File name: codingproblem5.ipynb
Author: Ji Yeol Yang
Date: August 17th, 2024
Version: 1.0
"""
# Import the random module for random selection of questions
import random

def initiate():
    """
    Prompts the user to enter the number of questions they would like to tackle.

    The function ensures that the user input is an integer within the valid range (1 to 10). 
    If the user gives a valid input, the function calls the `question_generator` function to start the quiz.
    If the user gives an invalid input, the function prints error messages and asks the user to try again.

    Input:
        User input that determines the number of quiz questions.

    Output:
        Calls the 'question_generator' function to prepare the questions for the quiz.

    Raises:
        ValueError: If the input is not a valid integer, an error message is displayed.
    """
    while True: # Start an indefinite while loop to ensure a valid input
        try:
            # Asks the user for the number of quiz questions
            num_of_questions = int(input('Enter in integer the number of questions you would like to tackle: '))

            # Ensures that the user gives an integer between 1 and 10
            if num_of_questions < 1 or num_of_questions > 10:
                print('You have to choose a number between 1 and 10!') # Error message to warn invalid range
            else:
                # If a valid input is given, call question_generator function with the user input passed as its argument
                question_generator(num_of_questions)
                break # Break the loop if valid input is given

        except ValueError:
            # Print a warning message if the input is not an integer and prompt the user to try again
            print('Invalid value. Please enter an integer.')


def question_generator(num_of_questions):
    """
    Prepares a set of random questions from the text file based on user input.

    This function reads multiple-choice questions and answers from a file, then randomly selects the 
    user-specified number of questions and passes them to the `quiz` function that starts the quiz.

    Input:
        num_of_questions (int): The number of questions to generate for the quiz.

    Output:
        Calls the `quiz` function that starts the quiz with a set of randomly selected questions.
    """
    question_dict = {} # Empty dictionary to store quiz questions and answers
    question = "" # Initialize empty string to concatenate questions
    answer = "" # Initialize empty string to concatenate answers

    # Open the file that stores the multiple-choice questions
    with open('mc_questions.txt', 'r') as r:
        # Loop through each line in the text file
        for line in r:
            line = line.strip() # Remove both unnecessary leading and trailing white spaces

            # If the first few characters do not start with 'Answer', it is part of the question
            if line[:6] != 'Answer':
                question += line + '\n' # Concatenate the question with the line
            # If the first few characters start with 'Answer', it is part of the answer    
            elif line[:6] == 'Answer':
                answer += line + '\n' # Concatenate the answer with the line
                question_dict[question] = answer # Line with answer signals an end of the question-answer pair. Store in dictionary
                question = "" # Reset question
                answer = "" # Reset answer

    # Randomly select the user-specified number of questions from the question dictionary
    random_questions = random.sample(list(question_dict.keys()), k = num_of_questions)
    
    # Call the quiz function to start the quiz by passing the dictionary and the randomly selected questions
    quiz(question_dict, random_questions)


def quiz(question_dict, random_questions):
    """
    Conducts the quiz by asking the user randomly selected questions. Displays and records the results.

    This function conducts the quiz by presenting the user with questions one by one. 
    It compares user answers to the correct answers and keeps track of any incorrect responses. 
    The results, including the user's answers and the correct ones, are displayed and written to a text file.

    Input:
        question_dict (dict): A dictionary containing questions as keys and answers as values.
        random_questions (list): A list of randomly selected questions.

    Output:
        Displays and writes the results to a text file and prints the user's score.
    """
    # Initialize an empty dictionary to store incorrect question-answer pairs
    incorrect_questions = {}

    # Create/overwrite a text file to record the quiz result
    with open('mc_result.txt', 'w') as w:
        w.write('Your Quiz result!\n\n') # Write the header
    
    # Loop through the randomly selected questions
    for i in range(len(random_questions)):
        quiz_question = random_questions[i] # Select the question one by one and store in the variable
        actual_answer = question_dict[random_questions[i]][8] # Store the corresponding answer from the dictionary pair

        # Display the question and prompt the user to answer
        print(random_questions[i])
        while True: # While loop to ensure that the user gives a valid input
            user_answer = input('Please enter your answer (a/b/c/d): ').lower()
            if len(user_answer) != 1: # This ensures the user does not enter the comma after the alphabet character
                print('Please enter a single alphabet character for your answer.\n')
            elif user_answer not in ['a', 'b', 'c', 'd']: # This ensures the input is from one of the four alphabets
                print('Enter a or b or c or d for your answer!\n')
            else:
                break # A valid input ends the loop
        
        # If the user provides an incorrect answer, the question and a list containing the correct answer and the user answer
        # are paired up and stored inside the incorrect_questions dictionary
        if actual_answer != user_answer:
            incorrect_questions[quiz_question] = [actual_answer, user_answer]

        # Open the same text file with the append mode        
        with open('mc_result.txt', 'a') as a:
            a.write(f'{quiz_question}\n') # Write the question
            a.write(f'Actual answer: {actual_answer}\n') # Write the actual answer
            a.write(f'Your answer: {user_answer}\n\n') # Write the user answer
            if actual_answer == user_answer: # Write if correct
                a.write('Correct!\n\n')
            else: # Write if incorrect
                a.write('Incorrect!\n\n')

    # Print and display the result of the quiz
    score = (len(random_questions) - len(incorrect_questions)) / len(random_questions) * 100 # Define the score
    if score == 100: # Print if perfect score
        print('Excellent Job! Perfect Score!')
    else: # Print if not perfect
        print(f'\nYour quiz score is {round(score, 2)}%.') # Print the score
        print('You got the following questions wrong: ') # Print the wrong questions
        for key, value in incorrect_questions.items(): # Loop through the incorrect questions dictionary
            print(key) # Print the question the user answered wrong
            print(f'Your answer: {value[1]}') # Print the user answer stored in index 1 of the list (value)
            print(f'Correct answer: {value[0]}') # Print the correct answer stored in index 0 of the list (value)

    # Open the same text file with the append mode
    with open('mc_result.txt', 'a') as a:
        if score == 100: # Write if perfect score
            a.write('Excellent Job! Perfect Score!')
        else: # Write if not perfect
            a.write(f'\nYour quiz score is {round(score, 2)}%.\n') # Write the score
            a.write('You got the following questions wrong:\n') # Write the wrong questions
            for key, value in incorrect_questions.items(): # Loop through the incorrect questions dictionary
                a.write(key) # Write the question the user answered wrong
                a.write(f'\nYour answer: {value[1]}\n') # Write the user answer stored in index 1 of the list (value)
                a.write(f'Correct answer: {value[0]}\n') # Write the correct answer stored in index 0 of the list (value)

if __name__ == '__main__':
    # Execute the initiate function when the script is run directly
    initiate()


```

### Question 1: Differences and Similarities Between Set and Dictionary

- *Similarities*:
    1. Both sets and dictionaries are mutable. This means that you can make changes to existing sets and dictionaries. You can add, delete, and update elements in a set or in a dictionary.
    2. Sets do not allow duplicate entries. Dictionaries do not allow duplicate keys.
    3. You can iterate over elements in sets and dictionaries.
    4. Both sets and dictionaries do not require specification of a fixed size upon creation. Their size is dynamic, meaning it adjusts as more elements are added or removed.
    5. Both sets and dictionaries are unordered, until the launch of Python 3.7 that allows dictionaries to have insertion order.
    6. Both sets and dictionaries store their elements by using the hash table. The hash table connects the elements in a set/dictionary to specific indices in an array.
- *Differences*:
    1. Sets do not allow duplicate elements. In dictionaries, although keys are unique, values CAN be duplicated.
        - The elements in a set must be immutable/hashable (integers, floats, strings, etc.). They cannot be mutable (lists, dictionaries, etc.).
        - The keys in a dictionary must be immutable/hashable (integers, floats, strings, etc.). The values of a dictionary can be both hashable and mutable, including lists and dictionaries.
    2. Structural difference: Sets contain unique elements. Dictionaries contain key-value pairs. An example of a set is `{'a','b','c'}`. An example of a dictionary is `{'a':1, 'b':2, 'c':3}`
    3. Their ideal usage is also different. While sets are preferred when picking and storing unique items, testing whether an element exists in a set (membership testing), and performing operations like unions, intersections, etc., dictionaries are preferred when storing and retrieving elements using the unique keys, looking up elements based on keys, storing structured data in key-value format, etc.  
- **Set:**
    - Usage: Sets are ideal for managing collections of distinct elements, conducting membership tests, and performing operations like unions, intersections, and differences.
        - Example Use Cases: 
            1. Staff Management: To maintain a unique list of employee IDs or names, ensuring no duplicates and allowing quick checks.
            2. Inventory Management: To track unique product IDs or names, easily identifying which items are in stock.
- **Dictionary:**
    - Usage: Dictionaries excel at storing and retrieving data using unique keys. They are efficient for managing structured data in key-value pairs.
        - Example Use Cases:
            1. Employee Information: To quickly access details about an employee using their unique ID, such as their profile or job role.
            2. Product Characteristics: To map product IDs to specific attributes, simplifying the retrieval and management of product information.
